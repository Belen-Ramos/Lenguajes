```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
datos = pd.read_csv("Escolaridad.csv")
datos.columns = ['Año', 'Sexo', 'Escolarización', 'Nivel']
datos
```

```python
datos.describe()
```

```python
filtroM = datos.loc[datos['Sexo'] == 'm']
filtroV = datos.loc[datos['Sexo'] == 'v']
```

```python
filtroM.describe()
#filtroV.describe()
```

```python
filtroMP = filtroM.loc[datos['Nivel'] == 'Primario']
PromMP = filtroMP['Escolarización'].mean()

filtroMM = filtroM.loc[datos['Nivel'] == 'Secundario / Medio']
PromMM = filtroMM['Escolarización'].mean()

filtroMS = filtroM.loc[datos['Nivel'] == 'Superior']
PromMS = filtroMS['Escolarización'].mean()
```

```python
filtroVP = filtroV.loc[datos['Nivel'] == 'Primario']
PromVP = filtroVP['Escolarización'].mean()

filtroVM = filtroV.loc[datos['Nivel'] == 'Secundario / Medio']
PromVM = filtroVM['Escolarización'].mean()

filtroVS = filtroV.loc[datos['Nivel'] == 'Superior']
PromVS = filtroVS['Escolarización'].mean()
```

```python
Array = np.array([PromMP, PromMM, PromMS, PromVP, PromVM, PromVS]).reshape(2,3)
datosNuevos = pd.DataFrame(Array, columns = ['Primario', 'Medio', 'Superior'], index = ['Femenino', 'Masculino' ])
datosNuevos
```

```python
fig, ax = plt.subplots()

Nivel = ['Primaria', 'Secundaria', 'Superior']
Masculino = [PromVP, PromVM, PromVS]
Femenino = [PromMP, PromMM, PromMS]

x = np.arange(len(Nivel))
ancho = 0.25

barraV = ax.bar(x-ancho/2, Masculino, ancho, label = 'Masculino')
barraM = ax.bar(x+ancho/2, Femenino, ancho, label = 'Femenino')

ax.set_ylabel('Escolarización')
ax.set_title('Nivel de escolarización promedio en CABA 2010-2017')
ax.set_xticks(x)
ax.set_xticklabels(Nivel)
ax.legend()

plt.savefig('doble_barra.png')
#Mostramos la grafica con el metodo show()
plt.show()
```
